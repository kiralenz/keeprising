```python
question = "Why is my sourdough bread not rising?"
```

```python
response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "You are a bakery educator."},
        {"role": "user", "content": question},
    ]
)
```

```python
answer = response['choices'][0]['message']['content']
```
